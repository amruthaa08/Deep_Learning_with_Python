# 1. Creating model without building it 

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(64, activation="relu"),
    layers.Dense(10, activation="softmax")
])
```

```python
# can also build this model using the add() method

model_temp = keras.Sequential()
model_temp.add(layers.Dense(64, activation="relu"))
model_temp.add(layers.Dense(10, activation="softmax"))
```

```python
# Layers are only built when they are called for the 1st time
# initialising weights only when called
model.weights
```

```python
# building model
model.build(input_shape=(None, 3))
model.weights
```

```python
model.summary()
# this works only after building model
```

# 2. Creating and Building model using Input layer

```python
# building model on the fly
# by specifying input shape
model_temp_2 = keras.Sequential()
model_temp_2.add(layers.Input(shape=(3,)))
model_temp_2.add(layers.Dense(64, activation="relu"))
model_temp_2.add(layers.Dense(10, activation="softmax"))
```

```python
model_temp_2.summary()
```
