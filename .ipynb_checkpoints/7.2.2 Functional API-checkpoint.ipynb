Sequential models can only be used when a single input and output is required.

```python
from tensorflow import keras
from tensorflow.keras import layers
```

# 1. Single input Single output

```python
# Creating same sequential model using Functional API
# go step by step from inputs
inputs = keras.Input(shape=(3,), name="my_input")   # symbolic tensor
features = layers.Dense(64, activation="relu")(inputs)
outputs = layers.Dense(10, activation="softmax")(features)

# instantiate model using Model constructor
# specify inputs and outputs
model = keras.Model(inputs=inputs, outputs=outputs)
```

# 2. Multi-Input, Multi-output

**three inputs:**
1. ticket title (text)
2. text body of ticket (text)
3. tags added by user (categorical)

**two outputs:**
1. priority score of ticket (scalar btwn 0 and 1) - sigmoid output
2. department to handle ticket (dept number) - softmax output

```python
# text as an array of 1s and 0s having length vocabulary_size
vocabulary_size = 10000
num_tags = 100
num_departments = 4

# creating input tensors
title = keras.Input(shape=(vocabulary_size,), name="title")
text_body = keras.Input(shape=(vocabulary_size,), name="text_body")
tags = keras.Input(shape=(num_tags,), name="tags")

# combining all individual inputs
features = layers.Concatenate(name="myconcatenate")([title, text_body, tags])
# intermediate layers
features = layers.Dense(64, activation="relu")(features)

# creating individual outputs
priority = layers.Dense(1, activation="sigmoid", name="priority")(features)
department = layers.Dense(num_departments, activation="softmax", name="department")(features)

# instantiating model
model = keras.Model(inputs=[title, text_body, tags],
                   outputs=[priority, department])
```

```python
model.summary()
```

```python
import numpy as np

# define fixed no of samples for our example
num_samples = 1280

# generating custom data
title_data = np.random.randint(0, 2, size=(num_samples, vocabulary_size))
text_body_data = np.random.randint(0, 2, size=(num_samples, vocabulary_size))
tags_data = np.random.randint(0,2, size=(num_samples, num_tags))

priority_data = np.random.random(size=(num_samples, 1))
department_data = np.random.randint(0, 2, size=(num_samples, num_departments))


# compiling model
model.compile(optimizer="rmsprop",
             loss=["mean_squared_error", "categorical_crossentropy"],
             metrics=[["mean_absolute_error"], ["accuracy"]])

# fitting mode
model.fit([title_data, text_body_data, tags_data],
         [priority_data, department_data],
         epochs=1)

# evaluating model
model.evaluate([title_data, text_body_data, tags_data],
              [priority_data, department_data])


# making predictions
priority_preds, department_preds = model.predict([title_data, text_body_data, tags_data])
```

# 4. Reusing layers + Feature extraction

**Feature extraction** - creating models that reuse intermediate features from another model

```python
model.layers    # provides list of layers
```

```python
model.layers[3].input
# get the input of a layer
```

```python
model.layers[3].output
# get the output of a layer
```

**Assumption**: You want to add another output to the previous model—you want to estimate how long a given issue ticket will take to resolve, a kind of difficulty rating. You
could do this via a classification layer over three categories: “quick,” “medium,” and
“difficult.” You don’t need to recreate and retrain a model from scratch. You can start
from the intermediate features of your previous model, since you have access to them

```python
# get the intermediate features from the 1st dense layer
features = model.layers[4].output

# create a layer to compute difficulty
difficulty = layers.Dense(3, activation="softmax", name="difficulty")(features)

# instantiate model again
new_model = keras.Model(
    inputs = [title, text_body, tags],
    outputs = [priority, department, difficulty]
)
```

```python
# plotting model
keras.utils.plot_model(
new_model, "images/updated_ticket_classifier.png", show_shapes=True)
```
