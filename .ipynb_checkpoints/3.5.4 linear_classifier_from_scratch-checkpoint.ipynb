# Building a linear classifier from scratch using our own synthetic data

- build 2 classes of data
- linearly seperable
- classes of points with their corrdinates

# Steps:
1. Create features and labels
2. Create model function (affine transformation)
3. Create loss function (mean squared error)
4. Create training loop

# 1. Creating data

#### numpy random multivariate_normal
The multivariate normal is a generalization of the one-dimensional normal distribution to higher dimensions. Such a distribution is specified by its mean and covariance matrix.
https://numpy.org/doc/stable/reference/random/generated/numpy.random.multivariate_normal.html
#### Covariance matrix
Describes shape of the point cloud. 
#### Mean 
Gives position of point cloud

```python
import numpy as np
import tensorflow as tf
```

```python
# define number of samples
num_samples_per_class = 1000

# create 2 classes of samples
neg_samples = np.random.multivariate_normal(
                mean=[0, 3],
                cov = [[1, 0.5], [0.5, 1]],
                size = num_samples_per_class)
pos_samples = np.random.multivariate_normal(
                mean=[3, 0],
                cov = [[1, 0.5], [0.5, 1]],
                size = num_samples_per_class)
```

```python
# stack them into single array of data
inputs = np.vstack((neg_samples, pos_samples)).astype(np.float32)
```

```python
inputs.shape
```

```python
# generate targets
targets = np.vstack((np.zeros((num_samples_per_class, 1)),
                   np.ones((num_samples_per_class, 1)))).astype(np.float32)
```

```python
targets.shape
```

```python
# plotting data with matplotlib
import matplotlib.pyplot as plt

# x = 1st col, y = 2nd col, color based on targets
plt.scatter(inputs[:, 0], inputs[:, 1], c = targets[:, 0])
plt.show();
```

# 2. Building linear classifier

```python
# give 2D coordinates as input and get class score as output
input_dims = 2
output_dims = 1

# weights and bias are tf Variables
# provide initial values 
W = tf.Variable(initial_value = tf.random.uniform(shape=(input_dims, output_dims)))
b = tf.Variable(initial_value = tf.zeros(shape=(output_dims,)))
```

# 3. Creating forward pass function

```python
# affine function y = Wx + b
def model(inputs):
    return tf.matmul(inputs, W) + b
```

# 4. Creating loss function

```python
def mean_squared_loss(true_vals, predicted_vals):
    per_sample_losses = tf.square(true_vals - predicted_vals)
    return tf.reduce_mean(per_sample_losses)
```

# 5. Creating Training loop

```python
# define constant learning rate
learning_rate = 0.1

def training_loop(inputs, targets):
    # instantiate gradient tape
    with tf.GradientTape() as tape:
        
        # find loss using predictions
        preds = model(inputs)
        loss = mean_squared_loss(targets, preds)
        
        
    # getting gradients
    grad_W, grad_b = tape.gradient(loss, [W, b])
    
    # updating weights
    W.assign_sub(grad_W * learning_rate)    
    b.assign_sub(grad_b * learning_rate)
    
    return loss
```

```python
# creating batch training loop
for step in range(40):    # 40 epochs
    loss = training_loop(inputs, targets)
    print(f"Loss at step {step} : {loss:.4f}")
```

```python
# plotting model performance
preds = model(inputs)
plt.scatter(inputs[:, 0], inputs[:, 1], c = preds[:, 0]>0.5)
plt.show()
```

```python
# plotting line fit to our data
x = np.linspace(-1, 4, 100)
y = -W[0] / W[1] * x + (0.5 - b) / W[1]
plt.plot(x, y, "-r")
plt.scatter(inputs[:, 0], inputs[:, 1], c = preds[:, 0]>0.5);
```
